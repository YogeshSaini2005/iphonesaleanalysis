#### top 10 high rate iphone on flipkart in india

```python
highest_Rated = data.sort_values(by=['Star Rating'], ascending = False)
highest_Rated = highest_Rated.head(10)
print(highest_Rated['Product Name'])
```

```python
print(highest_Rated)
```

#### how many rating highest rating do iphone have on flipkart in india

```python
data.head()
```

```python
iphones = highest_Rated['Product Name'].value_counts()
label =  iphones.index
counts = highest_Rated['Number Of Ratings']
figure = px.bar(highest_Rated, x = label, y = counts)
figure.show()
```

#### iphone highest no of reviews on flipkart 

```python
data.head()
```

```python
highest_reviews = data.sort_values(by=['Number Of Reviews'], ascending = False)
highest_reviews = highest_reviews.head(10)
print(highest_reviews['Product Name'])
```

```python
print(highest_reviews)
```

```python
iphones = highest_Rated['Product Name'].value_counts()
label =  iphones.index
counts = highest_Rated['Number Of Reviews']
figure = px.bar(highest_Rated, x = label, y = counts, title =" top rating iphones on flipkart in india with high rated")
figure.show()
```

#### relation between the sales price of iphone and no of rating on flipkart

```python
figure = px.scatter(data_frame = data , x = 'Number Of Ratings', y = 'Sale Price', size = 'Discount Percentage', trendline= 'ols', title= 'Relationship between Number of Ratings and Sale Price with Discount Percentage')
figure.show()
```
